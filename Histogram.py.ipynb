### Histogram

    Histogram are the most common means of looking at a single variable. 

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
df
```

```python
fig = plt.figure(figsize = (10,8))
axes1 = fig.add_subplot(1,1,1)
axes1.hist(df[['total_bill', 'tip']], color = ["gold", 'black'], orientation = "vertical")
axes1.set_title("Histogram of Total bills")
axes1.set_xlabel("Frequency")
axes1.set_ylabel("Total Bill")
```

    In the above program we find that histogram of total bill and tip values

```python
fig = plt.figure(figsize = (10,8))
axes1 = fig.add_subplot(1,1,1)
axes1.hist(df[['sex', 'smoker']], color = ["pink", 'green'], orientation = "vertical")
axes1.set_title("Histogram of sex")
axes1.set_xlabel("Frequency")
axes1.set_ylabel("smoker")
```

    In the above program we find that  histogram of gender values, about smoking male gender has highest smoking habit

```python
fig = plt.figure(figsize = (10,8))
axes1 = fig.add_subplot(1,1,1)
axes1.hist(df[['time', 'day']], color = ["red", 'gold'], orientation = "vertical")
axes1.set_title("Histogram of Day")
axes1.set_xlabel("Frequency")
axes1.set_ylabel("Time")
```

    In the above program find that according to time and day,in that time and day dinner is the most vising time and day is the saturday

```python
plt.figure(figsize = (2,2))
sns.countplot("day", data = df)
```

    In the above program find that at the saturday is the highest visiting day

```python
fig = plt.figure(figsize = (10,8))
axes1 = fig.add_subplot(1,1,1)
axes1.hist(df[['time', 'size']], color = ["red", 'gold'], orientation = "vertical")
axes1.set_title("Histogram of Time")
axes1.set_xlabel("size")
axes1.set_ylabel("Time")
```

    In the above code in the tips data lunch and dinner is the most visit timings
